```python
import sys, os

from time import time
import datetime
import numpy as np

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# Pandas parallelization
import swifter

sys.path.insert(0, os.path.abspath("./.."))
from ibgeapi import ibgeapi

# Import everything from utils file
from preprocessamento_mensal import *
```

```python
df = pd.read_parquet("../Scrapping Transparência/Data/data2010to2022.parquet")
display( df.head(3) )
```

```python
meses = np.array(list(np.arange(1,13)) * 13 )
anos = np.repeat(np.arange(2010, 2023), 12)
ipca_values = ibgeapi.consulta_ipca(meses, anos)

ipca = pd.DataFrame({"ano_exercicio": anos, "mes_referencia": meses, "ipca_value": ipca_values})
ipca.head(3)
```

```python
df_defl = df.merge(ipca, on = ["ano_exercicio", "mes_referencia"])

# Último valor de inflação 12/2022
ipca_ref = list(ipca.ipca_value)[-1]

df_defl["vl_despesa"] = df_defl.vl_despesa * ipca_ref / df_defl.ipca_value
df_defl = df_defl.drop(columns = "ipca_value")

display(df_defl.head(3))
```

```python
# Converte as informações de mes e ano para datas
df_defl = month_year_to_date(df_defl, year_name = "ano_exercicio", month_name = "mes_referencia")
df_defl.head(3)
```

```python
df_defl_quad = quadrimestral_agg(df_defl)
# df_defl_quad = date_to_quadrimestral(df_defl_quad)

df_defl_quad.head(3)
```

## Modelagem

```python
# Converte as categorias em códigos
df_defl_quad, le_dict = label_encoder(df_defl_quad)

# Conjunto de treino
df_defl_quad_train = df_defl_quad.loc[
    (df_defl_quad.data.dt.year >= 2018) & (df_defl_quad.data.dt.year < 2022), :
]
# Conjunto de teste
df_defl_quad_test = df_defl_quad.loc[
    df_defl_quad.data.dt.year == 2022, :
]

df_defl_quad_train_grouped = agrupar_dataframe(df_defl_quad_train)
df_defl_quad_test_grouped = agrupar_dataframe(df_defl_quad_test)
```

```python
padr = Padronizacao()
padr.fit_despesa(df_defl_quad_train_grouped)

padr.transform_despesa(df_defl_quad_train_grouped)
```

```python
df_defl_quad_train_scaled_grouped = padr.transform_despesa(df_defl_quad_train_grouped)
df_defl_quad_train_scaled = desagrupar_dataframe(df_defl_quad_train_scaled_grouped)

df_defl_quad_test_scaled_grouped = padr.transform_despesa(df_defl_quad_test_grouped)
df_defl_quad_test_scaled = desagrupar_dataframe(df_defl_quad_test_scaled_grouped)
```

```python
df_scaled_train = date_to_quadrimestral(df_defl_quad_train_scaled)
df_scaled_test = date_to_quadrimestral(df_defl_quad_test_scaled)

print("Dimensão Treino:", df_scaled_train.shape)
print("Dimensão Teste:", df_scaled_test.shape)

# df_scaled.head(6)
```

```python
x_train = df_scaled_train.loc[:,["ds_municipio", "ds_orgao", "id_despesa", "ano_exercicio", "quadrimestre"]]
y_train = df_scaled_train.loc[:,"vl_despesa"]
x_test = df_scaled_test.loc[:,["ds_municipio", "ds_orgao", "id_despesa", "ano_exercicio", "quadrimestre"]]
y_test = df_scaled_test.loc[:,"vl_despesa"]
```

```python
from skranger.ensemble import RangerForestRegressor
from lightgbm import LGBMRegressor
```

### Random Forests

```python
np.random.seed(42)

start = time.time()
rf = RangerForestRegressor(n_estimators = 100,
                           max_depth = 16,
                           quantiles = True,
                           importance = "impurity",
                           n_jobs = -1,
                           seed=42)
rf.fit(x_train, y_train)
print("Tempo de execução:", time.time() - start)
```

```python
y_train_pred_rf = rf.predict(x_train)
y_test_pred_rf = rf.predict(x_test)
y_train_pred_quantiles_rf = rf.predict_quantiles(x_train, [0.05, 0.95])
y_test_pred_quantiles_rf = rf.predict_quantiles(x_test, [0.05, 0.95])

df_scaled_train = df_scaled_train.assign(vl_despesa_pred_rf = y_train_pred_rf)
df_scaled_train = df_scaled_train.assign(vl_despesa_pred_inf_rf = y_train_pred_quantiles_rf[:,0])
df_scaled_train = df_scaled_train.assign(vl_despesa_pred_sup_rf = y_train_pred_quantiles_rf[:,1])
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_rf = y_test_pred_rf)
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_inf_rf = y_test_pred_quantiles_rf[:,0])
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_sup_rf = y_test_pred_quantiles_rf[:,1])
```

### LGBM

```python
params = {
    'objective': 'quantile',
    'metric': 'quantile',
    'max_depth': 16,
    'learning_rate': 0.01,
    'n_estimators': 200,
    'boosting_type': 'gbdt'
}

start = time.time()

lgb5 = LGBMRegressor(alpha = 0.05, **params)
lgb = LGBMRegressor(alpha = 0.5, **params)
lgb95 = LGBMRegressor(alpha = 0.95, **params)

model5 = lgb5.fit(x_train, y_train)
model = lgb.fit(x_train, y_train)
model95 = lgb95.fit(x_train, y_train)

print("Tempo de execução:", time.time() - start)
```

```python
y_train_pred_lgbm = model.predict(x_train)
y_test_pred_lgbm = model.predict(x_test)
y_train_pred5 = model5.predict(x_train)
y_test_pred5 = model5.predict(x_test)
y_train_pred95 = model95.predict(x_train)
y_test_pred95 = model95.predict(x_test)

df_scaled_train = df_scaled_train.assign(vl_despesa_pred_lgbm = y_train_pred_lgbm)
df_scaled_train = df_scaled_train.assign(vl_despesa_pred_inf_lgbm = y_train_pred5)
df_scaled_train = df_scaled_train.assign(vl_despesa_pred_sup_lgbm = y_train_pred95)
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_lgbm = y_test_pred_lgbm)
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_inf_lgbm = y_test_pred5)
df_scaled_test = df_scaled_test.assign(vl_despesa_pred_sup_lgbm = y_test_pred95)
```

```python
df_scaled_train = quadrimester_to_date(df_scaled_train, filtered_columns = ["data", "ds_municipio", "ds_orgao", "ds_elemento", "id_despesa", "vl_despesa",
                                                                            "vl_despesa_pred_rf", "vl_despesa_pred_inf_rf", "vl_despesa_pred_sup_rf",
                                                                            "vl_despesa_pred_lgbm", "vl_despesa_pred_inf_lgbm", "vl_despesa_pred_sup_lgbm"])
df_scaled_test = quadrimester_to_date(df_scaled_test, filtered_columns = ["data", "ds_municipio", "ds_orgao", "ds_elemento", "id_despesa", "vl_despesa",
                                                                            "vl_despesa_pred_rf", "vl_despesa_pred_inf_rf", "vl_despesa_pred_sup_rf",
                                                                            "vl_despesa_pred_lgbm", "vl_despesa_pred_inf_lgbm", "vl_despesa_pred_sup_lgbm"])
df_scaled = pd.concat([df_scaled_train, df_scaled_test])
df_scaled_grouped = agrupar_dataframe(
    df_scaled,
    list_columns = ["data", "vl_despesa", "vl_despesa_pred_rf",
                    "vl_despesa_pred_inf_rf", "vl_despesa_pred_sup_rf",
                    "vl_despesa_pred_lgbm", "vl_despesa_pred_inf_lgbm",
                    "vl_despesa_pred_sup_lgbm"]
)

df_scaled_grouped.head(3)
```

```python
df_grouped = padr.rescale_despesa(
    df_scaled_grouped,
    rescale_variables = [
        "vl_despesa", "vl_despesa_pred_rf", "vl_despesa_pred_inf_rf",
        "vl_despesa_pred_sup_rf", "vl_despesa_pred_lgbm", "vl_despesa_pred_inf_lgbm",
        "vl_despesa_pred_sup_lgbm"
    ]
)
```

```python
df_grouped = label_decoder(df_grouped, le_dict, categorical_columns = ["ds_municipio", "ds_orgao", "id_despesa"])
df_defl_quad = label_decoder(df_defl_quad, le_dict, categorical_columns = ["ds_municipio", "ds_orgao", "id_despesa"])
```
